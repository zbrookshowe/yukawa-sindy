```python
sim_list = ys.generate_training_data(mu_x0s=0.5, noise_level=0.1, scaled=True)

x_train = [sim.x for sim in sim_list[0:150]]
x_test = [sim.x for sim in sim_list[150:200]]
```

```python
threshold = 0.8
feature_names = ['x', 'v']
all_models = []
for i in range(10):
    opt = ps.STLSQ(threshold=threshold)
    weak_lib, strong_lib = ys.generate_libraries(sim_list[0].t)
    model = ps.SINDy(optimizer=opt, feature_library=weak_lib, feature_names=feature_names)
    all_models.append(copy(model))
    
```

```python
dummy_list = []
for model_elem in all_models:
    model_elem.fit(x_train, t=sim_list[0].t, multiple_trajectories=True)
    model_elem.print(precision=5)
    dummy_list.append(model_elem)

    score = model_elem.score(x_test, t=sim_list[0].t, multiple_trajectories=True,metric=root_mean_squared_error)
    print(f"score: {score:.5f}")
    print(60*'-' + '\n')
```
