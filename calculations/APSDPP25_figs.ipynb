```python
import pysindy as ps
import numpy as np
import matplotlib.pyplot as plt

import Yukawa_SINDy as ys
import cross_validation as cv
```

```python
sim_list = ys.generate_training_data(noise_level=0.1, scaled=True)
```

```python
sim_list[0].plot()
```

```python
example = sim_list[0]
pos = example.x[:,0]
vel = example.x[:,1]
t = example.t

fig, ax = plt.subplots()

# title and labels
ax.set_title("Sample noisy 2-particle trajectory")
ax.set_ylabel("Ptcl. separation ($\lambda_{Di}$)")
ax.set_xlabel("Time ($\omega_{pd}^{-1}$)")

# plot position
pos_curve = ax.plot(t, pos, lw=2, label='separation')

# plot velocity on right
ax_right = ax.twinx()
vel_curve = ax_right.plot(t, vel, '--k', lw=1, label='velocity')
ax_right.set_ylabel("Velocity ($\lambda_{Di}\omega_{pd}$)")
ax_right.set_ylim((0,3.))

# create legend
curves = pos_curve + vel_curve
labels = [x.get_label() for x in curves]
ax.legend(curves, labels)
```

```python
clean_pos = example.x_clean[:,0]
clean_vel = example.x_clean[:,1]

fig, ax = plt.subplots()

# title and labels
ax.set_title("Simulated 2-particle trajectory")
ax.set_ylabel("Ptcl. separation ($\lambda_{Di}$)")
ax.set_xlabel("Time ($\omega_{pd}^{-1}$)")

# plot position
pos_curve = ax.plot(t, clean_pos, lw=2, label='separation')

# plot velocity on right
ax_right = ax.twinx()
vel_curve = ax_right.plot(t, clean_vel, '--k', lw=1, label='velocity')
ax_right.set_ylabel("Velocity ($\lambda_{Di}\omega_{pd}$)")
ax_right.set_ylim((0,3.))

# create legend
curves = pos_curve + vel_curve
labels = [x.get_label() for x in curves]
ax.legend(curves, labels)
```

```python
noise = pos - clean_pos
fig, ax = plt.subplots()
ax.hist(noise, bins=20)
ax.set_xlabel("Position error ($\lambda_{Di}$)")
ax.set_ylabel("Counts")
ax.set_title("Added noise, $\sigma=0.1\lambda_{Di}$")
```
